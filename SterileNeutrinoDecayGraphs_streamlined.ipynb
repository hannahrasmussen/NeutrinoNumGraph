## Physical Constants

* ``alpha``: The unitless fine structure constant, $\alpha = 1/137$ 
* ``f_pi``: Pion decay constant, $f_{\pi} = 131 \text{ MeV}$
* ``Gf``: Fermi constant, $G_F = 1.166 \times 10 ^{-11} \text{ MeV}^{-2}$
* ``me``: Electron mass, $m_e = 0.511 \text{ MeV}$
* ``mpi_neutral``: Neutral pion mass, $m_{\pi^0} = 135 \text{ MeV}$
* ``mpi_charged``: Charged pion mass, $m_{\pi^{\pm}} = 139.569 \text{ MeV}$
* ``mu``: Muon mass, $m_{\mu} = 105.661 \text{ MeV}$

```python
a = 1/137         
f_pi = 131         
Gf = 1.166*10**-11 
me = .511          
mpi_neutral = 135  
mpi_charged = 139.569  
mu = 105.661
```

## Decay rates

### ``rate1``
* **Inputs**: 
    * ``ms``: Mass of the sterile neutrino 
    * ``angle``: Mixing angle of the sterile neutrino with the three active flavors
* **Outputs**: 
    * ``Gamma``: The decay rate of decay 1, $$\Gamma_1 = \frac{9 G_F^2 \alpha m_s^5 \sin^2(\theta)}{512 \pi^4}$$
    
### ``rate2``
* **Inputs**: 
    * ``ms``: Mass of the sterile neutrino 
    * ``angle``: Mixing angle of the sterile neutrino with the three active flavors
* **Outputs**: 
    * ``Gamma``: The decay rate of decay 2, $$\Gamma_2 = \frac{G_F^2 f_{\pi}^2}{16 \pi} m_s (m_s^2-m_{\pi^0}^2) \sin^2(\theta)$$
    
### ``rate3``
* **Inputs**: 
    * ``ms``: Mass of the sterile neutrino 
    * ``angle``: Mixing angle of the sterile neutrino with the three active flavors
* **Outputs**: 
    * ``Gamma``: The decay rate of decay 3, there is an extra 2 coefficient out front because $\nu_s$ can decay into either $\pi^+$ and $e^-$ OR $\pi^-$ and $e^+$, $$\Gamma_3 = 2 \frac{G_F^2 f_{\pi}^2}{16 \pi} m_s \sqrt{(m_s^2 - (m_{\pi^0}+m_e)^2)*(m_s^2 - (m_{\pi^0}-m_e)^2)} \sin^2(\theta)$$ 
    
### ``rate4``
* **Inputs**: 
    * ``ms``: Mass of the sterile neutrino 
    * ``angle``: Mixing angle of the sterile neutrino with the three active flavors
* **Outputs**: 
    * ``Gamma``: The decay rate of decay 4, there is an extra 2 coefficient out front because $\nu_s$ can decay into either $\pi^+$ and $\mu^-$ OR $\pi^-$ and $\mu^+$, $$\Gamma_4 = 2 \frac{G_F^2 f_{\pi}^2}{16 \pi} m_s \sqrt{(m_s^2 - (m_{\pi^{\pm}}+m_{\mu})^2)*(m_s^2 - (m_{\pi^{\pm}}-m_{\mu})^2)} \sin^2(\theta)$$ 
    
### ``rate5``
* **Inputs**: 
    * ``ms``: Mass of the sterile neutrino 
    * ``angle``: Mixing angle of the sterile neutrino with the three active flavors
* **Outputs**: 
    * ``Gamma``: The decay rate of decay 5, $$\Gamma_5 = \frac{G_F^2 m_s^5\sin^2(\theta)}{192\pi^3}$$

### ``rate6``
* **Inputs**: 
    * ``ms``: Mass of the sterile neutrino 
    * ``angle``: Mixing angle of the sterile neutrino with the three active flavors
* **Outputs**: 
    * ``Gamma``: The decay rate of decay 6, $$\Gamma_6 = \frac{\Gamma_5}{3}$$
    
### ``rate8``
* **Inputs**: 
    * ``ms``: Mass of the sterile neutrino 
    * ``angle``: Mixing angle of the sterile neutrino with the three active flavors
* **Outputs**: 
    * ``Gamma``: The decay rate of decay 8, $$\Gamma_8 = \frac{\Gamma_5}{3}$$
    
### ``ts``
* **Inputs**: 
    * ``ms``: Mass of the sterile neutrino 
    * ``angle``: Mixing angle of the sterile neutrino with the three active flavors
* **Outputs**: 
    * ``tau``: The overall lifetime of the sterile neutrino at the given mass and mixing angle in units of seconds, $$\tau_s = \frac{1}{\Gamma_1 + \Gamma_2 + \Gamma_3 + \Gamma_4 + \Gamma_5 + \Gamma_6 + \Gamma_8}$$
    
### ``decayrates``
* **Inputs**: 
    * ``ms``: Mass of the sterile neutrino 
    * ``angle``: Mixing angle of the sterile neutrino with the three active flavors
* **Outputs**: 
    * ``Gamma1``: The fraction of sterile neutrinos that decay via decay process 1, $\frac{\Gamma_1}{\sum\Gamma}$
    * ``Gamma2``: The fraction of sterile neutrinos that decay via decay process 2, $\frac{\Gamma_2}{\sum\Gamma}$
    * ``Gamma3``: The fraction of sterile neutrinos that decay via decay process 3, $\frac{\Gamma_3}{\sum\Gamma}$
    * ``Gamma4``: The fraction of sterile neutrinos that decay via decay process 4, $\frac{\Gamma_4}{\sum\Gamma}$
    * ``Gamma5``: The fraction of sterile neutrinos that decay via decay process 5, $\frac{\Gamma_5}{\sum\Gamma}$
    * ``Gamma6``: The fraction of sterile neutrinos that decay via decay process 6, $\frac{\Gamma_6}{\sum\Gamma}$
    * ``Gamma8``: The fraction of sterile neutrinos that decay via decay process 8, $\frac{\Gamma_8}{\sum\Gamma}$
    
### ``branching_ratios``
* **Inputs**: 
    * ``ms``: Mass of the sterile neutrino 
    * ``angle``: Mixing angle of the sterile neutrino with the three active flavors
* **Outputs**: 
    * ``ratio1``: The fraction of sterile neutrinos that decay via decay process 1, $\frac{\Gamma_1}{\sum\Gamma}$
    * ``ratio2``: The fraction of sterile neutrinos that decay via decay process 2, $\frac{\Gamma_2}{\sum\Gamma}$
    * ``ratio3``: The fraction of sterile neutrinos that decay via decay process 3, $\frac{\Gamma_3}{\sum\Gamma}$
    * ``ratio4``: The fraction of sterile neutrinos that decay via decay process 4, $\frac{\Gamma_4}{\sum\Gamma}$
    * ``ratio5``: The fraction of sterile neutrinos that decay via decay process 5, $\frac{\Gamma_5}{\sum\Gamma}$
    * ``ratio6``: The fraction of sterile neutrinos that decay via decay process 6, $\frac{\Gamma_6}{\sum\Gamma}$
    * ``ratio8``: The fraction of sterile neutrinos that decay via decay process 8, $\frac{\Gamma_8}{\sum\Gamma}$

```python
def rate1(ms,angle):  
    numerator = 9*(Gf**2)*a*(ms**5)*((np.sin(angle))**2)
    denominator = 512*np.pi**4
    Gamma = numerator/denominator
    return Gamma
    
def rate2(ms,angle):
    part1 = (Gf**2)*(f_pi**2)/(16*np.pi)
    part2 = ms*((ms**2)-(mpi_neutral**2))*(np.sin(angle))**2
    Gamma = part1*part2
    return Gamma

def rate3(ms,angle):
    part1 = (Gf**2)*(f_pi**2)/(16*np.pi)
    parentheses = ((ms**2) - (mpi_charged+me)**2)*((ms**2) - (mpi_charged-me)**2)
    part2 = ms * ((parentheses)**(1/2)) * (np.sin(angle))**2
    Gamma = part1*part2
    return 2*Gamma 

def rate4(ms,angle):
    part1 = (Gf**2)*(f_pi**2)/(16*np.pi)
    parentheses = ((ms**2) - (mpi_charged+mu)**2)*((ms**2) - (mpi_charged-mu)**2)
    part2 = ms * ((parentheses)**(1/2)) * (np.sin(angle))**2
    Gamma = part1*part2
    return 2*Gamma 

def rate4(ms,angle):
    part1 = (Gf**2)*(f_pi**2)/(16*np.pi)
    parentheses = ((ms**2) - (mpi_charged+mu)**2)*((ms**2) - (mpi_charged-mu)**2)
    part2 = ms * ((parentheses)**(1/2)) * (np.sin(angle))**2
    Gamma = part1*part2
    return 2*Gamma 

def rate5(ms,angle):
    numerator = (Gf**2)*(ms**5)*((np.sin(angle))**2)
    denominator = 192*np.pi**3
    Gamma = numerator/denominator
    return Gamma

def rate6(ms,angle):
    Gamma = (1/3)*(rate5(ms,angle))
    return Gamma

def rate8(ms,angle):
    Gamma = (1/3)*(rate5(ms,angle))
    return Gamma

def ts(ms,angle):
    tau_inv_MeV = 1/(rate1(ms,angle)+rate2(ms,angle)+rate3(ms,angle)+rate4(ms,angle)+rate5(ms,angle)+rate6(ms,angle)+rate8(ms,angle))
    tau_s = tau_inv_MeV*(6.582*10**-22)
    return tau_s

def decayrates(ms,angle):
    Gamma1= Gamma2= Gamma3= Gamma4= Gamma5= Gamma6= Gamma8 = 0
    if (ms>0):
        Gamma1 = rate1(ms,angle)
        Gamma5 = rate5(ms,angle) 
    if (ms>1.022): #the mass of two electrons
        Gamma6 = rate6(ms,angle)
    if (ms>211.32): #the mass of two muons
        Gamma8 = rate8(ms,angle)
    if (ms>mpi_neutral): #mass of a neutral pion
        Gamma2 = rate2(ms,angle)
    if (ms>(mpi_charged+me)): #mass of a charged pion plus the mass of an electron
        Gamma3 = rate3(ms,angle)
    if (ms>(mpi_charged+mu)): #mass of a charged pion plus the mass of a muon
        Gamma4 = rate4(ms,angle)
    return Gamma1, Gamma2, Gamma3, Gamma4, Gamma5, Gamma6, Gamma8

def branching_ratios(ms,angle):
    Gamma1, Gamma2, Gamma3, Gamma4, Gamma5, Gamma6, Gamma8 = decayrates(ms,angle)
    total = Gamma1 + Gamma2 + Gamma3 + Gamma4 + Gamma5 + Gamma6 + Gamma8
    ratio1 = Gamma1/total
    ratio2 = Gamma2/total
    ratio3 = Gamma3/total
    ratio4 = Gamma4/total
    ratio5 = Gamma5/total
    ratio6 = Gamma6/total
    ratio8 = Gamma8/total
    return ratio1,ratio2,ratio3,ratio4,ratio5,ratio6,ratio8
```

```python
mixangle = 5*10**-5
mass_arr1 = np.linspace(50,500,10000) #an array of possible masses of the sterile neutrino
process1 = np.zeros(len(mass_arr1))
process2 = np.zeros(len(mass_arr1))
process3 = np.zeros(len(mass_arr1))
process4 = np.zeros(len(mass_arr1))
process5 = np.zeros(len(mass_arr1))
process6 = np.zeros(len(mass_arr1))
process8 = np.zeros(len(mass_arr1))
for i in range (len(mass_arr1)):
    process1[i], process2[i], process3[i], process4[i], process5[i], process6[i], process8[i] = branching_ratios(mass_arr1[i],mixangle)
```

```python
from matplotlib import cm
plt.figure(figsize=(10,10))
plt.plot(mass_arr1,process1,color = 'gold', label="$\\nu_s \\to \\nu\gamma$")
plt.plot(mass_arr1,process2,color = 'salmon', linestyle =":", label="$\\nu_s \\to \\nu\pi^0$")
plt.plot(mass_arr1,process3,color = 'palevioletred', linestyle ="--", label="$\\nu_s \\to \pi^{\pm}e^{\mp}$")
plt.plot(mass_arr1,process4,color = 'rebeccapurple', linestyle ="-.", label="$\\nu_s \\to \pi^{\pm}\mu^{\mp}$")
plt.plot(mass_arr1,process5+process6+process8, color = 'indigo', label="Others")
plt.tick_params(axis="x", labelsize=20)
plt.tick_params(axis="y", labelsize=20)
plt.xlabel("Mass (MeV)", fontsize = 20)
plt.xticks(np.linspace(50,500,10))
plt.ylabel("Branching Ratio", fontsize = 20)
plt.legend(loc=0, fontsize=20)
plt.savefig("Branchingratio.pdf")
```

```python
n = 100
mass_arr2 = np.linspace(250,350,n)
mixangle_arr = np.linspace(1*10**-5,1*10**-4,n)
tau_matrix = np.zeros((n,n))

for i in range (n):
    for j in range (n):
        tau_matrix[i][j] = ts(mass_arr2[i],mixangle_arr[j])
        
time_array = [0.01,0.02,0.03,0.05,0.1,0.25,0.5,1.0,2.0]
colors = [cm.magma(x) for x in np.linspace(0.05, 0.8, len(time_array))]

fig1, ax1 = plt.subplots(figsize=(10,10))
CS1 = ax1.contour(mixangle_arr*10**5, mass_arr2, tau_matrix, time_array, colors = colors)
fmt = {}
strs = ['0.01 s','0.02 s','0.03 s','0.05 s','0.1 s','0.25 s','0.5 s','1.0 s','2.0 s']
for l, s in zip(CS1.levels, strs):
       fmt[l] = s

ax1.clabel(CS1, fmt=fmt, fontsize=18)
ax1.tick_params(axis="x", labelsize=20)
ax1.tick_params(axis="y", labelsize=20)
plt.xticks(np.linspace(2,10,5))
plt.xlabel("Mixing Angle ($\\times 10^{-5}$ Radians)", fontsize=20)
ax1.set_ylabel("Mass (MeV)", fontsize=20)
plt.savefig("DecayTimes.pdf")
```
